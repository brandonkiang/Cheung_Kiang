### Research Questions

Do colleges in the top quartile of academic selectivity (based on the incoming cohort's median ACT/SAT scores / acceptance rate) generate significantly different total athletic revenue than colleges in the bottom quartile?
* Rationale: This examines whether attracting highly competitive academic cohorts corresponds to a different scale of athletic financial operations.

How does the size of the incoming cohort moderate the relationship between total athletic recruiting expenses and total athletic revenue?
* Rationale: This explores the return on investment. Do schools bringing in massive freshman classes (often large state universities) see an exponential jump in athletic revenue for every dollar spent on recruiting, compared to schools bringing in small cohorts?

### Variables

Identify which variables to keep (which variables will be the most useful/informative to answer the research questions posed above)

* College related data variables (`UNITID`, `OPEID`, `INSTNM`, `CITY`, `STABBR`, `ZIP`)
* `ADM_RATE`: admissions rate at each campus
* `SATVR25`: SAT 25th percentile for reading section
* `SATVR75`: SAT 75th percentile for reading section
* `SATMT25`: SAT 25th percentile for math
* `SATMT75`: SAT 75th percentile for math
* `ACTCM25`: ACT composite median 25th percentile
* `ACTCM75`: ACT composite median 75th percentile
* `ADMCON7`: indicates if test scores are required, recommended, considered but not required, or neither required nor recommended. 
* `UGDS`: number of degree/certificate-seeking undergraduates enrolled in the fall
* `CONTROL`: public vs private
* `ICLEVEL`: 2-year vs 4-year


Additional Variables:
* `MAIN`: 1 indictes the institution is the main campus, 0 indicates it is not
* `NUMBRANCH`: number of branches at the institution
* `TUITIONFEE_IN`: Tuition and required fees for in-district students
* `TUITIONFEE_OUT`: Tuition and required fees for out-of-state students
* `NPT4_PUB`: average net price for public institutions
* `NPT4_PRIV`: average net price for private institutions
* `PCTPELL`: proportion of low income students

```python
cols = ['UNITID', 'OPEID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'SATVR25', 
        'SATVR75', 'SATMT25', 'SATMT75', 'ACTCM25', 'ACTCM75', 'ADMCON7', 'UGDS', 
        'CONTROL', 'ICLEVEL', 'MAIN', 'NUMBRANCH', 'TUITIONFEE_IN', 'TUITIONFEE_OUT', 
        'NPT4_PUB', 'NPT4_PRIV', 'PCTPELL']
df = pd.read_csv('MERGED2023_24_PP.csv', usecols=cols)

df = df.rename(columns={'UNITID': 'unitid', 'OPEID': 'opeid', 'INSTNM': 'institution_name', 'CITY': 'city', 'STABBR': 'state', 
                        'ZIP': 'zip', 'ADM_RATE': 'admission_rate', 'SATVR25': 'SAT_reading25','SATVR75': 'SAT_reading75', 
                        'SATMT25': 'SAT_math25', 'SATMT75': 'SAT_math75', 'ACTCM25': 'ACT_median25', 'ACTCM75': 'ACT_median75', 
                        'ADMCON7': 'test_score_req', 'UGDS': 'num_undergrad', 'CONTROL': 'institution_type', 'ICLEVEL': 'institution_years', 
                        'MAIN': 'main_campus', 'NUMBRANCH': 'num_campuses', 'TUITIONFEE_IN': 'in_state_tuition', 
                        'TUITIONFEE_OUT': 'out_state_tuition', 'NPT4_PUB': 'avg_net_price_public', 'NPT4_PRIV': 'avg_net_price_private', 
                        'PCTPELL': 'prop_low_income'})

df
```

```python
df["institution_type"] = df['institution_type'].map({1: 'public', 2: 'private nonprofit', 3: 'private for-profit'})
df
```

```python
df["institution_years"] = df['institution_years'].map({1: '4-year', 2: '2-year', 3: 'Less-than-2-year'})
df
```

```python
df['test_score_req'] = df['test_score_req'].map({1: 'Required', 2: 'Recommended', 3: 'Neither required nor recommended',
                                                 4: 'Do not know', 5: 'Considered but not required'})
df['test_score_req']
```

```python
df.isna().sum()
```

```python
df['institution_name'] = df['institution_name'].str.lower()
df['city'] = df['city'].str.lower()
df
```
