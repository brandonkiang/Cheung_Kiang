### Data Dictionary (Athletics)

- <b>unitid</b>: university ID

- <b>opeid</b>: Office of Postsecondary Education identification code

- <b>institution_name</b>: name of college or university

- <b>city</b>: City

- <b>state</b>: State Abbreviation

- <b>zip</b>: Zip Code

- <b>classification_name</b>: Division level of college athletics

- <b>EFTotalCount</b>: Total Athletes in the university

- <b>sector_name</b>: Public, Private, etc.

- <b>STUDENTAID_TOTAL</b>: Total Student Aid for all athletics teams

- <b>RECRUITEXP_TOTAL</b>: Total Recruiting expenses for all Athletics teams

- <b>GRND_TOTAL_REVENUE</b>: Grand Total Revenues

- <b>GRND_TOTAL_EXPENSE</b>: Grand Total Expenses

# Data Cleaning

Filter to only include the variables of interest identified earlier.

```python
athletics_columns = ["unitid", "opeid", "institution_name", "city", "state", "zip", "classification_name", "institution_type", "institution_years", "EFTotalCount", "STUDENTAID_TOTAL", "RECRUITEXP_TOTAL", "GRND_TOTAL_REVENUE", "GRND_TOTAL_EXPENSE"]
df_athletics = df_athletics[athletics_columns]
df_athletics.head()
```
